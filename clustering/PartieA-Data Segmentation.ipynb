# 2. Nettoyage et Pretraitement

```python
# Nettoyage et Pretraitement
dataset.dropna() # Supprimer les valeurs vides
```

## Matrice de Correlation en Scatter plot

```python
def replace_gender(value):
    if(value=='Male'):
        return 1
    else:
        return 0

print(dataset['Gender'])
dataset_modif=dataset['Gender'].map(lambda x:replace_gender(x))
dataset['Gender']=dataset_modif

print(dataset)

# corr_matrix = dataset.corr()
# print("La matrice de Coreelation est:",dataset)
# plt.scatter(corr_matrix)
```

```python
corr_matrix = dataset.corr()
print("La matrice de Coreelation est:",corr_matrix)
#plt.scatter(range(len(corr_matrix)),corr_matrix)
sns.scatterplot(corr_matrix)
```
